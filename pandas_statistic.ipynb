```python
import pandas as pd
```

# Exercise 1

```python
customer_df = pd.read_csv("adventure works 2020_customer.csv")
```

```python
customer_df['BirthDate'] = pd.to_datetime(customer_df['BirthDate'])
```

```python
#Calculation of customers age
customer_df['CustomerAge'] = (pd.to_datetime("today").year) - (customer_df["BirthDate"].dt.year)
customer_df['CustomerAge']
```

```python
#Split customers age into 4 span
bins = [1, 25, 40, 60, 100]
labels = [' < 25', '25 - 40', '40 - 60', ' >= 60']
customer_df['AgeSpan'] = pd.cut(customer_df['CustomerAge'], bins= bins, labels= labels, right = False)
```

```python
#Calculation of number of customers in 4 span age
customer_df.groupby('AgeSpan', as_index= False).agg(number = ('CustomerAge', 'count'))
```

There is no customer under age 25. **90%** customers are between **25 to 60** years old and gender of them are half male and half female.

# Exeercise 2

```python
sales_df = pd.read_csv('Adventure works 2020_sales.csv')
```

```python
sales_df['OrderDate'] = pd.to_datetime(sales_df['OrderDate'])
```

```python
#Finding weekday of every sale
sales_df['WeekDay'] = sales_df['OrderDate'].dt.day_name()
sales_df['WeekDay'].unique()
```

```python
#Calculation of sales base on weekdays
sales_df.groupby('WeekDay').agg(number = ('WeekDay', 'count')).sort_values(by = 'number', ascending= False)
```

Dispersion of number of sales is nearly close to each other. Most sales, 15%, are on the day **wednesday** during 2016 to 2019.

# Exercise 3

```python
product_df = pd.read_csv('adventure works 2020_product.csv')
```

```python
#Mask sales on 2018
sales_on_2018 = sales_df[sales_df['OrderDate'].dt.year == 2018]
sales_on_2018.head()
```

```python
sale_merge_with_product = sales_on_2018.merge(product_df[['ProductKey', 'ProductName']], on = 'ProductKey', how= 'left')
sale_merge_with_product.head()
```

```python
#Counting sales base on each product category
product_sales = sale_merge_with_product.groupby('ProductName')['OrderQuantity'].agg(order_quantity = 'sum').sort_values(by = 'order_quantity' , ascending= False)
product_sales.head(10)
```

**Water bottle** is in the first rank of order quantity. Products related to Bike are on the other ranks.

# Exercise 5 (method 1)

```python
sales_df['OrderDate'] = pd.to_datetime(sales_df['OrderDate'])
```

```python
#Identify last order of each customer
max_order_date = sales_df['OrderDate'].max()
max_order_date
```

```python
six_months_ago = max_order_date - pd.DateOffset(months= 6)
six_months_ago
```

```python
#Identify each customer doesnt purchase over past 6 month (6 month is calculated base on last date of order date (2019-07-31))
active_customers = sales_df[sales_df['OrderDate'] > six_months_ago]['CustomerKey'].unique()
active_customers
```

```python
inactive_customers = customer_df[~customer_df['CustomerKey'].isin(active_customers)]
inactive_customers.head()
```

```python
#Identification of inactive customers over past 6 months
six_month_inactive = inactive_customers[['CustomerKey', 'Name']].reset_index().drop(columns= 'index')
six_month_inactive.tail()
```

# Exercise 5 (method 2)

```python
#Identify last order of each customer
last_order = sales_df.groupby('CustomerKey')['OrderDate'].agg(last_order = 'max').reset_index()
last_order.head()
```

```python
#Identify each customer doesnt purchase over past 6 month (6 month is calculated base on last date of order date (2019-07-31))
inactive_customers = last_order[((sales_df['OrderDate'].max() - last_order['last_order']).dt.days) > 180]
inactive_customers.head()
```

```python
customer_merge_with_inactives = inactive_customers.merge(customer_df, on = 'CustomerKey', how = 'left')
customer_merge_with_inactives.tail()
```

```python
six_month_inactive = customer_merge_with_inactives[['CustomerKey', 'Name']].reset_index().drop(columns = 'index')
six_month_inactive.tail()
```

An analysis of **inactive customers** over the past six months revealed that out of total of 18,484 customers, 8,530 individuals ,equivalent to **46%**, had not made any purchase during this period. Among these inactive customers, approximately half were male and the other half were female.
